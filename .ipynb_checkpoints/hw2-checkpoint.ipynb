#### 1) A little introductory brain teaser. Which is more probable when rolling 2 six-sided dice: rolling snake eyes (two ones) or rolling sevens (dice sum to seven)? What is the ratio of the probabilities?

Knowing that there are multiple possibilities of getting a sum of seven, the probability of rolling two ones will be much less.

Probability of rolling two ones: $\frac{1}{6}*\frac{1}{6}=\frac{1}{36}$

Probability of rolling a sum of seven: $\frac{6}{36}$
    
    Possibilities: (1,6) (2,5) (3,4) (4,3) (5,2) (6,1)

Ratio of probabilites: $\frac{\frac{1}{36}}{\frac{6}{36}}=\frac{1}{6}$

#### 2) Following what we did in class show how to use the convolution operator to determine the probability of the sum of 2 six sided dice. Do both analytically (math & counting) and numerically (computer program). Beware the implicit definition of the values (x-axis on histogram).

```python
from scipy import stats, signal
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact

#---- Analytically

def roll(sum):
    p = 0
    for i in range(1,7):
        for j in range(1,7):
            if (i+j==sum):
                p += 1
    print("Analytical Probabilities: ", p/36)
interact(roll, sum=(2,12))

#---- Numerically

in1 = [1/6,1/6,1/6,1/6,1/6,1/6]
in2 = [1/6,1/6,1/6,1/6,1/6,1/6]
out = signal.fftconvolve(in1, in2)
print("Numerical Probabilities: ", out)
```

&nbsp;
#### 3) Calculate the mean and the variance of the distribution in problem 2. Hint: this is surprisingly tricky, make sure your result makes sense.


```python
#---- mean

print("\nMean: ", np.mean(out))

#---- variance

print("Variance: ", np.var(out))
```

#### 4) Repeat 2, and graph the average of 10 dice. Is this is a Gaussian distribution? Explain in depth.

```python
fig1, ax1 = plt.subplots(1,1)

out1 = signal.fftconvolve(in1, in2)
for n in range(1,10):
    out1 = signal.fftconvolve(out1, in1)
    
ax1.hist(np.arange(10,10+len(out1)), bins=len(out1), weights=out1);
# print("Standard deviation: ", stats.norm.ppf(out, max(np.arange(10,10+len(out)) * out)))
# print(np.arange(10,10+len(out))[np.where(out == max(out))])
```

This is Gaussian distribution because the standard deviation shows that 68% of the values fall within 1 standard deviation. There are an equal number of values below and above the mean as shown by the symmetry and the mean is equal to the median and mode.

#### 5) Show that the sum and average of an initially Gaussian distribution is also a Gaussian (can be analytic or numerical). How does the standard deviation of the resulting sum or average Guassian change? This is a hugely important result. Explore what this means for integrating a signal over time.

```python
fig2, (ax1,ax2,ax3) = plt.subplots(1,3,figsize=(15,5))

# standard

print("\nNormal:  ", np.std(stats.norm.pdf(out1)))

# summing

print("Sum:     ", np.std(5*stats.norm.pdf(out1)))

# averaging

out2 = out1
for n in range(1,11):
    out2 = signal.fftconvolve(out2, in1)
print("Average: ", np.std(out2))
    
# samples = []
# for n in range(1,10000):
#     samples.append(np.random.choice(np.arange(10, 66), p=out))
# print(np.std(samples))

ax1.plot(np.linspace(stats.norm.ppf(0.01), stats.norm.ppf(0.99), len(out1)), out1);
ax2.plot(np.linspace(stats.norm.ppf(0.01), stats.norm.ppf(0.99), len(out1)), 5*out1);
ax3.plot(np.arange(10, 10+len(out2)/2, 0.5), out2);
```

The standard deviation varies greatly but the shape of the distribution is still the same. For integrating a signal over time, the spread of the data will change over an observation interval.
